```python
import pandas as pd
import numpy as np
import json 
import requests
from io import StringIO
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
from math import ceil
import datetime
import calendar
```

```python
# Specify the path to your JSON configuration file
config_file_path = "config.json"

# Load the JSON configuration file
with open(config_file_path, 'r') as config_file:
    config= json.load(config_file)

# Print the configuration
#Connection_STRING = config["connectionString"]


CONNECTION_STRING_AZURE_STORAGE = config["connectionString"]
CONTAINER_AZURE= 'airbnb'

# Initialize the BlobServiceClient
blob_service_client= BlobServiceClient.from_connection_string(CONNECTION_STRING_AZURE_STORAGE )

# Get the container client
container_client = blob_service_client.get_container_client(CONTAINER_AZURE)


airbnb_df= pd.DataFrame()

# List all blobs in the specified container
blob_list = container_client.list_blobs()
for blob in blob_list:
    print(blob.name)
    blob_client = container_client.get_blob_client(blob=blob.name)
    blob_data = blob_client.download_blob()
    blob_content = blob_data.readall().decode('utf-8')  # Corrected line
    df = pd.read_csv(StringIO(blob_content))
    # Display the shape of the DataFrame
    print(df.shape)
    # Since there's only one CSV, I'm going to assign it to airbnb_df
    airbnb_df = df.copy()
```

```python
airbnb_df['last_review'] = pd.to_datetime(airbnb_df['last_review'])
airbnb_df.head()
```

```python
# Filter out rows where the number_of_reviews is lower than 10
airbnb_df = airbnb_df[airbnb_df['number_of_reviews'] >= 10]

# Filter out rows where the year of last review is not 2024
airbnb_df = airbnb_df[airbnb_df['last_review'].dt.year == 2024]

# Filter out rows where availability_365 is not equal to 0
airbnb_df = airbnb_df[airbnb_df['availability_365'] != 0]

# Filter out rows where number_of_reviews_ltm is not equal to 0
airbnb_df = airbnb_df[airbnb_df['number_of_reviews_ltm'] != 0]

airbnb_df.head()
```

```python
dim_host_df= airbnb_df[['host_id', 'host_name']].drop_duplicates(subset= 'host_id')
dim_host_df.head()
```

```python
airbnb_df['room_type_id'] = pd.factorize(airbnb_df['room_type'])[0] + 1
airbnb_df.head()
```

```python
dim_room_df= airbnb_df[['room_type_id', 'room_type']].drop_duplicates(subset= 'room_type_id')
dim_room_df.head()
```

```python
airbnb_df['location_id'] = pd.factorize(airbnb_df['neighbourhood'])[0] + 1
airbnb_df.head()
```

```python
dim_location_df= airbnb_df[['location_id', 'neighbourhood_group' , 'neighbourhood', 'longitude', 'latitude']].drop_duplicates(subset= 'location_id')
dim_location_df.head(40)
```

```python
dim_facts_df = airbnb_df[['minimum_nights','number_of_reviews', 'location_id', 'price', 'host_id', 'room_type_id']]
dim_facts_df.head()
```
